```python
def display_misclassification_matrices(confusion_matrices):
    """
    Displays only misclassifications for each confusion matrix individually with the same color scale.
    Args:
        confusion_matrices (dict): Dictionary of confusion matrices with each value being a list of lists or np.array.
    """
    # Plot each confusion matrix with only misclassifications displayed
    for matrix_name, matrix in confusion_matrices.items():
        # Convert to numpy array and zero out the diagonal (correct classifications)
        matrix = np.array(matrix)
        np.fill_diagonal(matrix, 0)

        # Create a new figure for each matrix
        plt.figure(figsize=(6, 6))
        sns.heatmap(matrix, annot=True, cmap="Reds", fmt="d", vmin=0, vmax=20)
        plt.title(f"{matrix_name} Misclassifications")
        plt.xlabel("Predicted Label")
        plt.ylabel("True Label")

        # Display the plot
        plt.show()

# Call the function to display only misclassifications
display_misclassification_matrices(confusion_matrices)
```

```python
def aggregate_misclassifications(confusion_matrices, threshold=0.1, name="All"):
    """
    Aggregates misclassifications across multiple confusion matrices of potentially varying dimensions.

    Args:
        confusion_matrices (dict): Dictionary of confusion matrices with each value being a list of lists or np.array.
        threshold (float): The minimum misclassification ratio to consider significant.

    Returns:
        aggregated_misclassifications (dict): Dictionary with total and average misclassifications across matrices.
    """
    # Determine the maximum number of classes across all matrices
    max_classes = max(len(matrix) for matrix in confusion_matrices.values())
    print("Max classes = ", max_classes)

    # Initialize arrays to store aggregated counts and ratios
    total_misclassifications = np.zeros((max_classes, max_classes), dtype=int)
    misclassification_ratios = np.zeros((max_classes, max_classes), dtype=float)
    matrix_count = len(confusion_matrices)

    # Iterate over each confusion matrix
    for matrix_name, matrix in confusion_matrices.items():
        matrix = np.array(matrix)  # Ensure it's a numpy array
        n_classes = matrix.shape[0]
        total_instances = matrix.sum(axis=1)

        for i in range(n_classes):
            for j in range(n_classes):
                if i != j:
                    total_misclassifications[i, j] += matrix[i, j]
                    misclassification_ratios[i, j] += matrix[i, j] / total_instances[i]

    # Calculate average misclassification ratios
    avg_misclassification_ratios = misclassification_ratios / matrix_count

    # Print significant misclassifications based on threshold
    print(f"Significant Misclassifications Across {name} Matrices:".format(threshold))
    for i in range(max_classes):
        for j in range(max_classes):
            if i != j and avg_misclassification_ratios[i, j] > threshold:
                print(f"Class {i} misclassified as {j}: Total count = {total_misclassifications[i, j]}, "
                      f"Average Ratio = {avg_misclassification_ratios[i, j]:.2f}")

    # Visualization
    plt.figure(figsize=(12, 6))

    # Heatmap of total misclassifications
    plt.subplot(1, 2, 1)
    sns.heatmap(total_misclassifications, annot=True, cmap="YlOrRd", fmt="d")
    plt.title(f"Total Misclassifications Across {name} Matrices")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")

    # Heatmap of average misclassification ratios
    plt.subplot(1, 2, 2)
    sns.heatmap(avg_misclassification_ratios, annot=True, cmap="YlOrRd", fmt=".2f")
    plt.title(f"Average Misclassification Ratios Across {name} Matrices")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")

    plt.tight_layout()
    plt.show()

    # Return aggregated data for further analysis if needed
    return {
        "total_misclassifications": total_misclassifications,
        "avg_misclassification_ratios": avg_misclassification_ratios
    }
```
